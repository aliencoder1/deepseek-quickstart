```python
import os
import base64

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

base64_image = encode_image("1141765162709_.pic_hd.jpg")
```

## Grounding

```python
completion = client.chat.completions.create(
    model="qwen3-vl-plus",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:image/png;base64,{base64_image}"}, 
                },
                {"type": "text", "text": "请输出图中所有蓝色沙发的位置，输出格式为物体检测框，输出为json。"},
            ],
        }
    ],
)
print(completion.choices[0].message.content)
```

## Final Prompt

```python
completion = client.chat.completions.create(
    model="qwen3-vl-plus",
    messages=[
        {
            "role": "user",
            "content": [
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:image/png;base64,{base64_image}"}, 
                },
                {"type": "text", "text": """请完成如下任务：
1、输出图中所有蓝色沙发的位置，输出格式为物体检测框。
2、请统计图中总共有多少个蓝色沙发。只需要输出个数，
3、描述图中蓝色沙发。

输出结果格式：
{
    "grounding": {},
    "counting": {},
    "vqa": {}
}

只需要输出json，不需要有其他输出。
"""},
            ],
        }
    ],
)
print(completion.choices[0].message.content)
```
